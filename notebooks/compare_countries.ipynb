```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

# 1. Load Cleaned Data
countries = ['ethiopia', 'kenya', 'sudan', 'somalia', 'djibouti']
all_data = []

for c in countries:
    # Adjust path to where your clean CSVs are stored
    df = pd.read_csv(f"../data/{c}_clean.csv") 
    all_data.append(df)

df_all = pd.concat(all_data, ignore_index=True)

# 2. Temperature Summary Table (KPI Requirement)
temp_summary = df_all.groupby('Country')['T2M'].agg(['mean', 'median', 'std']).reset_index()
print("Temperature Summary Statistics:\n", temp_summary)

# 3. Statistical Testing (ANOVA KPI)
# Testing if temperature differences between countries are significant
groups = [df_all[df_all['Country'] == c]['T2M'].dropna() for c in countries]
f_stat, p_val = stats.f_oneway(*groups)
print(f"\nANOVA Result: p-value = {p_val:.4f}")

# 4. Extreme Heat Days (Days > 35°C)
extreme_heat = df_all[df_all['T2M_MAX'] > 35].groupby('Country').size().reset_index(name='Heat_Days')
sns.barplot(data=extreme_heat, x='Country', y='Heat_Days')
plt.title("Extreme Heat Days (T2M_MAX > 35°C)")
plt.show()
```
